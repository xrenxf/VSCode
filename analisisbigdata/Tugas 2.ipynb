### Step 8. Tampilkan item yang paling banyak diorder (gunakan groupBy(...))?

```python
gd = data.groupby('item_name')['quantity'].sum()
most_ordered_item = gd.sort_values(ascending=False)
display(most_ordered_item)
```

### Step 9. Untuk item terbanyak diorder, berapa item yang diorder?

```python
grouped_data = data.groupby('item_name')['quantity'].sum()
most_ordered_item = grouped_data.sort_values(ascending=False).head(1)
display(most_ordered_item)
```

### Step 10. Apa isi kolom choice_description pada item terbanyak diorder ini?

```python
grouped_data = data.groupby('item_name')['quantity'].sum()
most_ordered_item_name = grouped_data.sort_values(ascending=False).index[0]
most_ordered_item = data[data['item_name'] == most_ordered_item_name]
print(most_ordered_item[['item_name', 'choice_description']])
```
